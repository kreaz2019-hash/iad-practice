# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as dss
```

## Load the Data

```python
wine = dss.load_wine()
```

```python
type(wine)
```

```python
wine
```

```python
wine['feature_names']
```

```python
feature_names = wine['feature_names'][:5]
feature_names
```

```python
wine['target']
```

```python
# create data from the first 5 columns
df = pd.DataFrame(
    data = wine['data'][:, :5],
    columns = feature_names
)

df.info()
```

```python
# add a column
df['target'] = wine['target']
df.info()
```

# Pairwise correlation of columns

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(5, 4)) # size of the box
sns.heatmap(corr,  cmap='coolwarm', annot=True, fmt=".2f")
plt.show()
```

## Boxplots
for more examples see https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
for i in range(len(feature_names)):
    figure = plt.figure()
    ax = sns.boxplot(x='target', y=feature_names[i], data=df)
```

## Pairplot 
for more examples see 
https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(df, hue="target", palette = 'Set1')
```

```python
# 
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 4)


from sklearn.datasets import load_wine
wine = load_wine(as_frame=True)
df = wine.frame
```

```python
# 1)    (pairplot)
pp = sns.pairplot(df, hue="target", corner=True, diag_kind="kde", plot_kws=dict(alpha=0.8, s=25))
pp.fig.suptitle("Wine: Pairplot ( vs )", y=1.02)
plt.show()
```

```python
# 2) Violin/Box   
for col in ["alcohol", "flavanoids", "color_intensity", "proline"]:
    plt.figure()
    sns.violinplot(data=df, x="target", y=col, cut=0, inner="quartile")
    plt.title(f"{col}   (Wine)")
    plt.xlabel("target"); plt.ylabel(col)
    plt.show()

```

```python
# 3)    
plt.figure(figsize=(9, 7))
sns.heatmap(df.drop(columns=["target"]).corr(), cmap="vlag", center=0, square=True)
plt.title("Wine:   ")
plt.show()
```
